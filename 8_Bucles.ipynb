# Break out // Rompiendo un loop
image.png

```python
n = 0
while True:
    if n == 3:   #1 2 3
        break
    print(n)   #0 1 2
    n = n + 1  #1 2 3
    
print('off')
  
```

# Terminando una iteraci√≥n con Continue

image.png

```python
while True:
  line = input('> ')
  if line[0] == '#':
      continue
  if line == 'done':
      break
  print(line)
print('Done')
```
