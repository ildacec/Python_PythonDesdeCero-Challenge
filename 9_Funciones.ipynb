```python
#definir con retorno y 2 par√°metros 
def sumar(a,b):
  total = a + b
  return total

#llamado 
print(sumar(15,5))
print(sumar('15','5'))

suma = sumar(15,5)

if suma > 100:
  print('valor mayor que 100')
else:
  print ('valor menor que 100')
```

```python
#definir con retorno y 2 par√°metros 
def oper_aritmeticas(a,b):
  suma = a + b
  print(suma)
  resta = a - b
  print(resta)
  multipli = a * b
  print(multipli)
  divi = a // b
  print(divi) 


oper_aritmeticas(25,12)
```
